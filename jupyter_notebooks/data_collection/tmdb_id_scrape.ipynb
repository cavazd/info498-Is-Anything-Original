```python
import pandas as pd
import requests 
import time 
```

```python
def get_movie_by_id(id):
    api_key = '716978917b3565e92ac410463f95122c'
    url = f"https://api.themoviedb.org/3/find/{id}?external_source=imdb_id&api_key={api_key}"
    res = requests.get(url)
    res = res.json()['movie_results'][0]
    # hope the first value is the movie we are looking for
    return {
            'original_language': res['original_language'],
            'original_title': res['original_title'],
            'overview': res['overview'],
            'popularity': res['popularity'],
            'release_date': res['release_date'],
            'vote_average': res['vote_average'],
            'vote_count': res['vote_count']
        }
```

```python
# Read the list from the text file
with open("5-24_incomplete_ids/incomplete_scraper_ids_5-24.txt", "r") as file:
    loaded_list = file.read().splitlines()

print(loaded_list)
```

```python
all_movies = []
failed_ids = []
for id in loaded_list:
    time.sleep(0.4)
    try:
        movie_info = get_movie_by_id('tt'+ id)
        all_movies.append(movie_info)
    except:
        print("Error: " + id)
        failed_ids += id
    

print("Collected " + len(all_movies) + " movies successfully. " + len(failed_ids) + " errors.")
```

```python
all_movies_df = pd.DataFrame(all_movies)
all_movies_df
```

```python
all_movies_df.to_csv('tmdb_full_info.csv', index=False)
```
