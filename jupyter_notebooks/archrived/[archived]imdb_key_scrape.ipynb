## Movie Collection

### Get Movie IDs Based on Keyword Search

```python
# GET FULL DATA SET
keyword = "remake"

page_items = 1
search = []
i = 1
while (page_items > 0):
    new_collect = ia.get_keyword(keyword, page=i)
    search.extend(new_collect)
    page_items = len(new_collect)
    i += 1
    print("Retrieved page", i)

search_id = [title.movieID for title in search]
print("Collected", len(search_id), "items over", i-2, "pages")
print(search_id)
```

### Prep methods for collection

```python
# Storing type value for Person object
test = ia.get_movie('0380510', ['main', 'vote details'])
personType = type(test.get('producer')[0])

# Method takes a list of Person or Company objects and returns it as a list of dicts with an ID and name
def normalize_objects(objects):
    new_list = []
    for item in objects:
        if type(item) is personType:
            id = item.personID
        else:
            id = item.companyID
        name = item.data['name']
        new_list.append({
            'id': id,
            'name': name
        })
    return new_list
```

```python
# Returns information for media based on inputted ID if media is a movie. If not, returns None and prints that it was removed
def get_movie_info(id):
    res = ia.get_movie(id)
    ia.update(res, 'keywords')
    
    if (res['kind'] == "movie"):
        return {'imdbID': id,
               'title': res.get('title'),
               'production companies': normalize_objects(res.get('production companies')),
               'director': normalize_objects(res.get('director')),
               'producer': normalize_objects(res.get('producer')),
               'genres': res.get('genres'),
               'keywords': res.get('keywords'),
               'rating': res.get('rating'),
               'votes': res.get('votes')
               }
    else:
        print("Removed - Not a Movie: ", ia.get_movie(id)['title'])
    
```

### Collect Movies 

```python
import requests
import time 
```

```python
imdb_data = []
error_counter = 0
id_list = search_id[:5]
for id in id_list:
    time.sleep(0.4)
    try:
        movie_info = get_movie_info(id)
    except:
        error_counter += 1
        print('Error occurred:', ia.get_movie(id)['title'])
        movie_info = {'imdbID': None,
               'title': None,
               'production companies': None,
               'director': None,
               'producer': None,
               'genres': None,
               'keywords': None,
               'rating': None,
               'votes': None
               }
    imdb_data.append(movie_info)

# Removes empty rows from imdb_data
imdb_data = [item for item in imdb_data if item is not None and item['imdbID'] is not None]

print("Completed with", error_counter, "Errors and", (len(id_list) - len(imdb_data) - error_counter), "Removed Value(s)")
```
